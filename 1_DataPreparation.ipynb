# 2 DataFrame Preparation & Manipulation

## 2.1 Weather Stations DataFrame Manipulation

We are going to fetch just wind meassurements at Weather Stations (Wind, Wind Gust, Wind direction) that will be the target variables to be determined with the features from the historicalweather dataset. The time granularity at historical weather dataset is 3h so we are going to use the same in the actual wind dataset. 

***In the first iteration we are going to collect just actual wind data(target) and just for the target location (tarifa, SPAIN, lon = -5.59883, lat=36.013985  ), but might consider collect data from other stations in further iterations that can help to improve the predictions.***

```python
weatherStation_tarifa = pd.read_csv('Data/StationsRecords/DH-6001.csv', sep = ';',
                 header = None,
                 usecols=[5, 11,10,14]
                )
```

```python
colnames = {5:'DateTime',
            11:'Tarifa_Wind',
            10:'Tarifa_WindGust',
            14:'Tarifa_WindDirection',
            }

weatherStation_tarifa.rename(columns=colnames, inplace=True)
```

```python
weatherStation_tarifa.head()
```

```python
weatherStation_tarifa.isna().sum()
```

**Things to be taken ino account:**

- **Granularity** : We want to adapt the granularity of Actuals(1h) to Historical Forecast (3h). For that purpose we are going to use the mean values of wind, gust and direction every three hours.

*A brief analysis between addapting the granuarity of the actuals to the granuarlity of the forecasts or viceversa. If we would have decide to addapt the granularity of forecasts (3h) to the actuals one(1h), we would have to interpolate values, then creating non-real values. Conversely, if we addapt actuals (1h) to forecast (3h), we would do a mean of 3 real actual meassurements, what is a real value.*


- **Direction of the wind:** Wind direction is given in degrees so it has to be taken into account that 360º and 0º are the same angle (N). A transformation need to be done to allow the model to understand that they are similar values.  

The strategy will be:

**STEP 1** Addapt granularity ot 3 h.

**STEP 2** Wind linear component to allow Machine Learning Model to better undestand the wind direciton (359º similar to 1º)

**STEP 3** Wind Speed in knots, that suits better in nautical environment. 





 - **STEP 1 : GRANULARITY**   Mean values for Wind and Wind Gust every 3 hours (granularity)

```python
%%time

for i in weatherStation_tarifa.index:
    
    if int(weatherStation_tarifa['DateTime'][i][-8:-6])%3==0:
        weatherStation_tarifa.loc[i,'Tarifa_Wind'] = weatherStation_tarifa.loc[i:i+2,'Tarifa_Wind'].mean()
        weatherStation_tarifa.loc[i,'Tarifa_WindGust'] = weatherStation_tarifa.loc[i:i+2,'Tarifa_WindGust'].mean()
    
    else:
        weatherStation_tarifa.drop(i, axis=0, inplace=True)

weatherStation_tarifa.reset_index(drop=True, inplace=True)
```

```python
weatherStation_tarifa
```

```python
weatherStation_tarifa.isna().sum() # We still have a small percetenge of NaN to be treated later
```

- **STEP 2 WIND COMPONENTS** To allow Machine Learning models to treat the  wind direction, angles degree shall be descomposed into linear components N-S and E-W (the ML model wont be able to interpret that 359 º and 1 º are quite similar directions if we dont do this tranformation). 

```python
# Function that return the linear component of an angular direction

def wind_components(wind_direction):
    NS_component = np.cos(np.radians(wind_direction))
    EW_component = np.sin(np.radians(wind_direction))
    return NS_component, EW_component
```

```python
(weatherStation_tarifa['Tarifa_Direction N-S'],
 weatherStation_tarifa['Tarifa_Direction E-W']) = wind_components(weatherStation_tarifa['Tarifa_WindDirection'])

weatherStation_tarifa
```

```python
weatherStation_tarifa
```

- **STEP 3 - WIND SPEED & WIND GUST IN KNOTS** as is the standard in nautical domains.

```python
# Function that receive wind in m/s and return it in knots

def wind_knots(speed):
    return speed * 1.94384
```

```python
weatherStation_tarifa['Tarifa_Wind'] = wind_knots(weatherStation_tarifa['Tarifa_Wind'])
weatherStation_tarifa['Tarifa_WindGust'] = wind_knots(weatherStation_tarifa['Tarifa_WindGust'])

weatherStation_tarifa
```

 ## 2.2 Historical Forecast DataFrame Preparation

For the purpose of this project and in order to simplify the data, we are going to maintain just one forecast for each "slice dt". The forecast selected will be the one made in the same day by 00.00 h that should be very accurate. 

***In other words, the forecast made by 00.00 hour will be kept as a prediction for the next 24 hours. The prediction interval for slice dt prediction for the first 24 hours is 3 hours, so we will maintain (24h/3h) 8 slice dt per forecast***

*Additionally* 
- We are going to kept just forecast from 6th of Feb 2019, that is the first record available at DataStations Datasets.
- Wind speed & gust in knots will be converted from m/s to knots
- Wind direction components N-S and E-W will be added

The strategy would be:

- **STEP 1** - Create a DataFrame with forecast dt from 2019-02-06 and on
- **STEP 2** - Filter forecast dt made by 00.00h
- **STEP 3** - From the previous DataFrame, keep just the first 8 slice dt (24h) from each forecast dt made by 00.00 hours
- **STEP 4** - Wind in knots
- **STEP 5** - Wind N-S & E-O components added

**STEP 1** - Create a dataset with forecast dt from 2019-02-06 and on

```python
df_chunk = pd.read_csv('Data/historicalForecast.csv', chunksize=10000)

chunk_list = []

for chunk in df_chunk:
    chunk_list.append(chunk)

    
historicalForecast = pd.concat(chunk_list)
```

```python
historicalForecast.shape
```

```python
pd.set_option('display.max_rows', 10)

resultado = historicalForecast['forecast dt iso'].str.find('2019-02-06 00:00:00 +0000 UTC')
first_coincidence = list(resultado[resultado == 0].index)[0] 
```

```python
# first coincidence in row 180420

historicalForecast = historicalForecast[historicalForecast.index>=first_coincidence]
historicalForecast.reset_index(drop=True, inplace=True)
```

```python
historicalForecast
```

**STEP 2** -Filter forecast dt made daily by 00.00 h

```python
# 1 Filter forecast dt made by 00.00h

historicalForecast = historicalForecast[historicalForecast['forecast dt iso'].str.contains('00:00:00')]
historicalForecast.reset_index(drop=True, inplace=True)
```

**STEP 3** - Keep just the first 8 slice dt (24h) from each forecast dt made by 00.00 h

```python
%%time
# 2 keep just the first 8 slice dt forecasts (24h) for each forecast dt by 00.00h

aux = historicalForecast.iloc[0:8,:]

for i in range (9,len(historicalForecast)):
    if historicalForecast['forecast dt iso'][i] == historicalForecast['slice dt iso'][i] :
        aux = pd.concat([aux,historicalForecast.iloc[i:i+8,:]])
        
historicalForecast = aux
```

```python
historicalForecast.reset_index(drop=True, inplace=True)
historicalForecast
```

- **STEP 4** - Wind in knots


```python
historicalForecast['wind_speed'] = wind_knots(historicalForecast['wind_speed'])
```

- **STEP 5** - Wind N-S & E-W components added


```python
(historicalForecast['Direction N-S'], historicalForecast['Direction E-W']) = wind_components(historicalForecast['wind_deg'])
```
